## 1. Definición

Los Números de Lucas se definen como:
\begin{gather*}
  L_n :=
    \begin{Bmatrix}
      2 \hspace{2.5cm} & \hspace{1cm} si \hspace{0.5cm} n=0 \\
      1 \hspace{2.5cm} & \hspace{1cm} si \hspace{0.5cm} n=1 \\
      L_{n-1} + L_{n-2} & \hspace{1cm} si \hspace{0.5cm} n>1
    \end{Bmatrix}
\end{gather*}


```
# Definición de los Números de Lucas
def lucas(n):
  if n == 0:
    return 2
  if n == 1:
    return 1
  return lucas(n-1) + lucas(n-2)
```

## 2. Número de Lucas más cercano a 1000

Este punto se realiza en dos pasos. Primero, se calculan los dos números más próximos a 1000: el número más próximo por abajo se guarda en `arr[2]` y su índice $n$ se guarda en `arr[0]`. El número más próximo por arriba se guarda en `arr[3]` y su índice en `arr[1]`:



```
# Número de Lucas más cercano a 1000
## Paso 1: Numero más próximo por debajo y por encima:
def closeThousandArray():
  arr = [0,0,0,0]
  n = 0
  while lucas(n) < 1000:
      n += 1
  arr[0] = n-1
  arr[1] = n
  arr[2] = lucas(n-1)
  arr[3] = lucas(n)
  return arr
```

Posteriormente, se elige el más próximo a 1000 entre los dos. Para ello, se determina la separación entre cada uno de los números y 1000:

```
## Paso 2: Número más próximo entre los dos:
def closeThousand():
  arr = closeThousandArray()
  abs1 = 1000 - arr[2]
  abs2 = arr[3] - 1000
  c = [0,0]
  if abs1 < abs2:
    c[0] = arr[0]
    c[1] = arr[2]
    return c;
  else:
    c[0] = arr[1];
    c[1] = arr[3];
    return c
```

## 3. Primer Número de Lucas mayor que 100

Se comienza con $L_0 = 2$. Luego, se va incrementando el índice $n$ de $L_n$ mientras $L_n < 100$. Esta condición deja de cumplirse cuando encontramos en Número de Lucas mayor a 100. Este sería el número buscado:

```
# Primer Número de Lucas mayor que 100
def greaterHundred():
  n = 2
  while lucas(n) < 100:
      n += 1
  arr[0] = n;
  arr[1] = lucas(n)
  return arr;
```
